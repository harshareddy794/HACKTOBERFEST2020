## Import Modules

```python
import pandas as pd # To read csv
import re # Regex
import seaborn as sb # Statistical data visualization
import matplotlib.pyplot as plt
```

## Data Preparation and Cleaning

```python
df = pd.read_csv('Pokemon.csv')
df.head()
```

```python
df.columns = ['#','Name','Type1', 'Type2','Total' ,'HP', 'Attack','Defense','Sp.Atk','Sp.Def','Speed', 'Generation', 'Legendary']
```

```python
df.head(10)
```

**Duplicate text in front of the Pokemon name needs to be removed**

```python
df.Name = df.Name.apply(lambda x: re.sub(r'(.+)(Mega.+)',r'\2',x))
df.Name = df.Name.apply(lambda x: re.sub(r'(.+)(Primal.+)',r'\2',x))
df.Name = df.Name.apply(lambda x: re.sub(r'(HoopaHoopa)(.+)','Hoopa'+r'\2',x))
```

**Removing Mega form and Primal forms of the pokemons** 

```python
All = df.loc[(df['Name'].str.contains('Mega ')==False) & (df['Name'].str.contains('Primal')==False)] # filter all pokemon without Mega and Primal in name.
All
```

**Dropping Legendary and Generation col**

```python
new_all_df = All.drop(['Generation','Legendary'], axis=1)
```

```python
new_all_df.head()
```

**Creating a new df which shows the average of each stats for each individual types**

```python
types_stats = new_all_df.melt(id_vars=['Total' ,'HP', 'Attack','Defense','Sp.Atk','Sp.Def','Speed'],
                value_vars=['Type1','Type2'],
                value_name='Type').drop('variable', axis=1).dropna(axis=0)
types_stats = types_stats.groupby('Type').mean()
```

```python
types_stats
```

**Conclusion**: From the above df we can see that Dragon Type pokemons on average have highest total stats.

## Asking and Answering Questions

### Q1) Which type of pokemon on average is the most powerful ?

```python
# Sorting the df in descending order
types_stats = types_stats.sort_values('Total',ascending=False)
```

### Barplot : Avg total stat vs Types

**Conclusion**: From the above graph we can see that overall the **Dragon Type** is most powerful, followed by Steel and  Psychic.

### Q4) What are the top 5 pokemons, each of different types?

```python
def max_stat(t):
    df_type = All[(All['Type1']==t) & (All['Legendary']==False)]
    return df_type.loc[df_type['Total'].idxmax()]
```

max_stat is a function which returns the best pokemon for a given type, provided the pokemon is not a legendary.

```python
t1_types = All['Type1'].unique()
l = []
for t in t1_types:
    l.append(max_stat(t))
df_top = pd.concat(l, axis=1, keys=[s.name for s in l]).T.drop(['Type2', 'Generation', 'Legendary'], axis=1)
df_top = df_top.reset_index().drop('index',axis=1)
df_top = df_top.sort_values('Total', ascending=False)
```

### Q5) Find the top 10 Pokemon in a specific Generation, by a given statistic.

```python
def Top10(Gen,Bstat):
        subGen = All[All['Generation']==Gen]
        Strong = subGen.sort_values([Bstat], ascending = False).drop(['Generation', 'Legendary', '#'],axis=1)
        print(Strong[['Name','Type1','Type2',Bstat]].head(10))
        return
```
